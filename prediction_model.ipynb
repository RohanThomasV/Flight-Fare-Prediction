```python

import pandas as pd


```

```python
data=pd.read_excel("Data_Train.xlsx")
data.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
```

```python
data["Source"]=l.fit_transform(data["Source"])
```

```python
data["Airline"]=l.fit_transform(data["Airline"])
data["Destination"]=l.fit_transform(data["Destination"])
data["Total_Stops"]=l.fit_transform(data["Total_Stops"])
data["Additional_Info"]=l.fit_transform(data["Additional_Info"])
```

```python
data.head(10)
```

```python
data.drop(["Date_of_Journey"],axis=1,inplace=True)
```

```python
 
 
data.drop(["Route"],axis=1,inplace=True)
```

```python
data.head(10)
```

```python
 
data.drop(["Dep_Time"],axis=1,inplace=True)
 
```

```python
data.head(10)
```

```python
data["Duration"]=l.fit_transform(data["Duration"]) 
```

```python
data.head(10)
```

```python
x=data[['Price']]
```

```python
y=data.drop(['Price'],axis=1)
```

```python
x
```

```python
y
```

```python
 y
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(ytrain,xtrain)
```

```python
model.predict(ytest)
```

```python
ytest
```

```python
xtest
```

```python
model.score(ytest,xtest)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
modell=LinearRegression()
```

```python
modell.fit(ytrain,xtrain)
```

```python
modell.predict(ytest)
```

```python
modell.score(ytest,xtest)
```
